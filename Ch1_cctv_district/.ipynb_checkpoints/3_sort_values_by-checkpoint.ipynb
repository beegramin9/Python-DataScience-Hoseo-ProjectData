```python
''' pandas 이용해서 CCTV와 인구 현황 데이터 파악하기 
1. CCTV 현황 '''
import pandas as pd
CCTV_Seoul = pd.read_csv('CCTV_in_Seoul.csv')
print(CCTV_Seoul.head()) #  그냥 생으로 파일을 가져옴
print(CCTV_Seoul.sort_values(by='소계', ascending = True).head(5)) # CCTV가 적은 수 부터
CCTV_Seoul.sort_values(by='소계', ascending = False).head(5) # CCTV가 많은 수 부터
```

```python
''' 최근 3년간의 증가율을 계산하기 '''
CCTV_Seoul['최근 증가율'] = (
        CCTV_Seoul['2014년']+
        CCTV_Seoul['2015년']+
        CCTV_Seoul['2016년']
)/ CCTV_Seoul['2013년도 이전'] * 100

print(CCTV_Seoul.sort_values(by='최근 증가율', ascending = False).head(5))
CCTV_Seoul.sort_values(by='최근 증가율', ascending = True).head(5)
```

```python
''' 2. 인구 현황 '''
pop_Seoul = pd.read_excel('Population_in_Seoul.xls',
                        header = 2,
                         usecols = 'B,D,G,J,N')
pop_Seoul.rename(columns = {
    pop_Seoul.columns[0] : '구별',   # Column에서의 인덱싱이 필요하니까
    pop_Seoul.columns[1] : '인구수', # Column[0] 이렇게 써 줘야겠지?
    pop_Seoul.columns[2] : '한국인',
    pop_Seoul.columns[3] : '외국인',
    pop_Seoul.columns[4] : '고령자'
}, inplace = True) # 실제 Dataframe의 내용을 바꾸라는 얘기

pop_Seoul.head()
```

```python
''' 총 합계(맨 처음 행)는 필요가 없습니다'''
# pop_Seoul.drop([0], inplace = True) # 1.행 2.열 이니까 행의 인덱스를 갈 땐 rows라고 안하고
                                    # 열의 인덱스를 찾을 때만 columns 라고 합니다.
''' [0] 이 없다고 나와있는데 왜 그렇죠?
이미 inplace = True를 때렸으니 0이 없어진 상태로 바뀌어버린거죠!'''

''' 데이터가 중복된 게 있을까요? unique 판별 '''
print(pop_Seoul['구별'].unique())

''' 끝에 NaN이 보이는데, 이 NaN이 어디에 있는 지 확인하기 위해서는 조건문을 사용해야 합니다.
리스트 안에 조건문을 넣으면 인덱스가 나오죠?'''
pop_Seoul[pop_Seoul['구별'].isnull()] 
```

```python
''' 그러면 이제 NaN이 있는 행을 없애야겠죠? '''
# pop_Seoul.drop([26],inplace = True)
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100 
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100 
```
