```python
''' CSV 파일 읽기 '''
import pandas as pd # 모든 라이브러리는 클래스다!
                    # 그러므로 모든 함수는 타입함수(메소드) .head('인수 있으면 여기') 형태

CCTV_Seoul = pd.read_csv('CCTV_in_Seoul.csv', encoding = 'UTF-8') # pd클래스의 메소드
CCTV_Seoul.head() # 처음 5행까지만 보여준다.
```

```python
''' 열 인덱스를 읽고 싶어요.'''
CCTV_Seoul.columns # 얘는 왜 메소드가 아닌 지 모르겠네. 
# CCTV_Seoul.columns[0]
```

```python
''' "기관명" 이름을 "구별" 이름으로 바꾸고 싶어요! '''
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]: '구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
''' 엑셀 파일 읽기 '''
pop_Seoul = pd.read_excel('Population_in_Seoul.xls')
pop_Seoul.head()
```

```python
''' 2번째 행까지 헤더가 있죠? 그리고 읽고싶은 열만 읽을게요! '''
pop_Seoul = pd.read_excel('Population_in_Seoul.xls',
                         header = 2,
                          usecols = 'B,D,G,J,N' # 책에서는 parse_cols라고 나와있는데, 이제는 usecols 를 사용합니다.
                         )
pop_Seoul.head()
```

```python
''' 이름이 계, 계.1... 이렇게 나와있어서 알아보기 힘들다.
바꿔야지'''
pop_Seoul.rename(columns ={
    pop_Seoul.columns[0] : '구별'
    pop_Seoul.columns[1] : '인구수'
    pop_Seoul.columns[2] : '한국인'
    pop_Seoul.columns[3] : '외국인'
    pop_Seoul.columns[4] : '고령자'
}, inplace = True)
```
