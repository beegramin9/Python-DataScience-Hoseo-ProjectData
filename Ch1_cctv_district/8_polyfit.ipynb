```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0,10,0.01)
y = 3*t + 5

plt.figure( figsize = (12,8))
plt.plot(t,y)
plt.xticks(color = 'w')
plt.yticks(color = 'w')

plt.show()
```

```python
'''        !'''
y_noise = y + np.random.rand(len(y))
#  len    
plt.figure( figsize = (12,8))
plt.plot(t, y_noise)
plt.xticks(color = 'w')
plt.yticks(color = 'w')
plt.show()
```

```python
'''    
polyfit     . '''
fp1 = np.polyfit(t, y_noise, 1)
print(fp1)
'''      (    )
(function)   
poly1d : A one-dimensional polynomial class 
  polynomial class   .       
   .
'''
f1 = np.poly1d(fp1)
f1
```

```python
''' polynomial class :    numpy  '''
print(np.poly1d([1,1])  + np.poly1d([1,-1]))
np.poly1d([1,1])  * np.poly1d([1,-1])
```

```python
plt.figure( figsize = (15,12) )
plt.plot(t, y_noise , label = 'noise', color = 'y')
plt.plot(t, y , ls = 'dashed', lw=3 , color = 'b', label ='original')
plt.plot(t, f1(t),  lw = 2 , color = 'r', label = 'polyfit')
plt.xticks(color = 'w')
plt.yticks(color = 'w')
# f1        f1(t)
plt.grid()
plt.legend()
plt.show()
```
