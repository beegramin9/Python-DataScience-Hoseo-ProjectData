# 서울시 구별 주유 가격 확인하기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import folium
import googlemaps
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)
```

```python
stations = pd.read_csv('04. stations.csv', sep =',', encoding = 'utf-8')
stations.sort_values(by = '가격', ascending = True, inplace = True)
del stations['Unnamed: 0']
stations.head(10)
```

```python
stations.sort_values(by = '가격', ascending = False, inplace = True)
stations.head(10)
```

- 이것만으로는 알아보기 힘들지? 구 별로 pivot_table 해 봅시다!

```python
gu_data = pd.pivot_table(stations, index = ['구'], values = ['가격'], aggfunc = np.mean)

gu_data
```

```python
geo_json = '04. skorea_municipalities_geo_simple.json'
geo_data = json.load(open(geo_json, encoding = 'utf-8'))

st_map = folium.Map(location = [37.5502, 126.982] , zoom_start = 11,
                   tiles= 'Stamen Toner')
```

```python
st_map.choropleth(geo_data = geo_data,
                  data = gu_data,
                  columns = (gu_data.index, '가격'),
                  key_on = 'id', # 여기에 뭐가 들어갈지는 직접 Json 파일을 보고 결정한다.
                  fill_color = 'PuRd'
)
st_map
```

# 서울시 주유 가격 상하위 10개 지도에 표시하기

```python
stations.sort_values(by = '가격', ascending = False, inplace = True)
oil_price_top10 = stations.head(10)
oil_price_top10
```

- 주소 가져오기

```python
oil_price_top10['주소']
```

```python
from tqdm import tqdm
lat = []
lng = []

for add in tqdm(oil_price_top10['주소']):
    try:
        tmp_add = add.split("(")[0]
        tmp_map = gmaps.geocode(tmp_add , language = 'ko')
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print('One NaN spotted')
    
oil_price_top10['latitute'] = lat
oil_price_top10['longitute'] = lng
oil_price_top10
```

```python
stations.sort_values(by = '가격', ascending = True, inplace = True)
oil_price_bot10 = stations.head(10)
oil_price_bot10
```

```python
lat = []
lng = []

for add in tqdm(oil_price_bot10['주소']):
    try:
        tmp_add = add.split("(")[0]
        tmp_map = gmaps.geocode(tmp_add , language = 'ko')
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print('One NaN spotted')
    
oil_price_bot10['latitute'] = lat
oil_price_bot10['longitute'] = lng
oil_price_bot10
```

```python
st_map = folium.Map(location = [37.5202, 126.975] ,zoom_start = 10.5)

for n in oil_price_top10.index:
    if pd.notnull(oil_price_top10['latitute'][n]):
        folium.CircleMarker(location = [oil_price_top10['latitute'][n],oil_price_top10['longitute'][n]],
                            radius = 15, color = '#CD3181', fill_color = '#CD3181'
                           ).add_to(st_map)
        
for n in oil_price_bot10.index:
    if pd.notnull(oil_price_bot10['latitute'][n]):
        folium.CircleMarker(location = [oil_price_bot10['latitute'][n],oil_price_bot10['longitute'][n]],
                            radius = 15, color = '#3186cc', fill_color = '#3186cc'
                           ).add_to(st_map)

st_map
```
