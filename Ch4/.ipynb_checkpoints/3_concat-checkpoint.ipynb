```python
import pandas as pd
from glob import glob
```

### 파일들의 리스트를 뽑을 떄 사용하는 glob
- 정규표현식을 사용해서

```python
glob('./stations/지역*.xls')
```

```python
station_files = glob('./stations/지역*.xls')
```

```python
test = pd.read_excel('./stations/지역_위치별(주유소).xls', header =2 )
test.head(10)
```

```python
tmp_raw = []

for file in station_files:
    tmp = pd.read_excel(file, header = 2)
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
station_raw
```

```python
stations = pd.DataFrame({
    '이름':station_raw['상호'],
    '주소':station_raw['주소'],
    '가격':station_raw['휘발유'],
    '셀프':station_raw['셀프여부'],
    '상표':station_raw['상표']
})
stations
```

```python
stations['구'] = [ eachname.split()[1] for eachname in stations['주소']]
stations
```

### 데이터가 너무 많아! Unique() 검사

```python
stations['구'].unique()
```

```python
stations[stations['구']=='강동구']
```

### 가격이 제대로 기입되지 않은 곳이 있다?

```python
stations[stations['가격']=='-']
```

- 일일히 확인 할 수는 없으니 제외하도록 합시다

```python
stations = stations[stations['가격']!='-']
stations.head(15)
```

```python
stations['가격']
```

```python
stations['가격'] = [ float(value) for value in stations['가격'] ]
```

- 25개의 엑셀을 합쳤으므로 index가 중복될 수 있습니다. 그래서 reset_index로 처음부터 인덱싱을 다시!

```python
stations.reset_index(inplace = True)
stations
```

- 근데 이러면 Index 열이 새로 생겨버리니까 지워버리자

```python
del stations['index']
```
