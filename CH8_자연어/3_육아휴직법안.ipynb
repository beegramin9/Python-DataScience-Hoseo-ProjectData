```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
import warnings
warnings.filterwarnings("ignore")
```

```python
import nltk
from konlpy.tag import Okt, Kkma, Hannanum
```

```python
from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read() # 육아휴직법안 파일 ID
```

```python
doc_ko[:1000]
```

```python
# 명사 뽑아오기
okt = Okt()
tokens_ko = okt.nouns(doc_ko)
tokens_ko[:10]
```

```python
# 사용된 총 단어의 종류(개수)
ko = nltk.Text(tokens_ko, name = '대한민국 국회 의안 제 1809890호')
len(ko.tokens), len(set(ko.tokens))
```

```python
# 단어의 빈도수
ko.vocab()
```

```python
# 빈도수 그래프
plt.figure(figsize = (15,8))
ko.plot(50)
plt.show()
```

- 한국어는 불용어를 일일히 입력해야 한다

```python
stop_words = ['.','(',')',',',"'",'%','-','X',').','x','의','자','에','안','번','호','을','이','다','만','로','가','를']

# 불용어 제거
ko = [each_word for each_word in ko if each_word not in stop_words]
ko[:10]
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')
plt.figure(figsize=(15,6))
ko.plot(50) # 50개만 보기
plt.show()
```

```python
# 초등학교가 몇 번 나왔을까?
ko.count('초등학교')
```

```python
# 단어의 대략적인 위치
plt.figure( figsize= (15,6) )
ko.dispersion_plot(['육아휴직','초등학교','공무원'])
```

```python
# 해당 단어 주위에 어떤 단어들이 있는지
ko.concordance('초등학교')
```

```python
# 연어(Collocations), 책에는 collations()라 나와있지만 collocation_list()로 써야함
ko.collocation_list()
```

```python
from wordcloud import WordCloud
data = ko.vocab().most_common(150)

wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                     relative_scaling=0.2,
                     background_color='white',
                     width=1000,height=600).generate_from_frequencies(dict(data))
            
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
