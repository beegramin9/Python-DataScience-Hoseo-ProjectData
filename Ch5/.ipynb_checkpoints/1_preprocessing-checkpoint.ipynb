# 우리나라 인구 소멸 위기 지역 분석
우리나라니까 서울시가 아니라 대한민국 전역 지도가 그려져야 합니다.

65세 이상 노인 인구와 20 ~ 39세 여성 인구를 비교하여,
20 ~ 39세 여성/65세 이상 노인 인구 < 0.5 일 때 인구 소멸 위험 지역

국가통계포털 https://kosis.kr/index/index.do

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import platform
from matplotlib import rc, font_manager
%matplotlib inline

path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family = font_name)

''' 혹시 마이너스가 있을 때는? '''
plt.rcParams['axes.unicode_minus'] = False
```

```python
pop = pd.read_excel('05. population_raw_data.xlsx', header = 1)
pop
```

```python
''' 엑셀이 많이 복잡해서 전처리를 좀 합시다. 
결측값을 열 기준 위에서 아래방향으로 채우기. 그래야지 도시별로 이쁘게 묶이겠지? '''

pop.fillna(method='pad', inplace = True)
pop.head(5)
```

```python
pop.rename(columns = {
    '행정구역(동읍면)별(1)':'광역시도',
    '행정구역(동읍면)별(2)':'시도',
    '계':'인구수'
},inplace= True)
pop.head(5)
```

```python
''' 소계는 전국, 시 전체 인구를 합해 버리니까 쪼개서 볼 수 없음
어짜피 구로 쪼개서 밑에 나 나와있으니 차라리 날려버려도 됨'''

pop = pop[ (pop['시도'] != '소계')]
pop.head()
```

```python
''' 항목 부분의 이름이 너무 거창하지? loc를 사용해서 바꿔버리자 '''
pop.rename(columns = {
    '항목':'구분'
},inplace = True)

pop.loc[pop['구분'] == '총인구수 (명)', '구분'] = '합계'
pop.loc[pop['구분'] == '여자인구수 (명)', '구분'] = '여자'
pop.loc[pop['구분'] == '남자인구수 (명)', '구분'] = '남자'
pop.head()
```

# 기준에 따라 20 ~ 39, 65세 이상으로 표 정리하기

```python
pop['20 - 39세'] = pop['20 - 24세'] + pop['25 - 29세'] + pop['30 - 34세'] + pop['35 - 39세']

pop['65세 이상'] = pop['65 - 69세'] + pop['70 - 74세'] + pop['75 - 79세'] + pop['80 - 84세'] + \
                   pop['85 - 89세'] + pop['90 - 94세'] + pop['95 - 99세'] + pop['100+'] 
```

```python
pop.head()
```

Pivot_table, values와 columns

```python
pop = pd.pivot_table(pop , index =['광역시도','시도'],
                     columns = ['구분'], # values 값들을 추가로 또 구별하고 싶을 때
                    values = ['인구수', '20 - 39세', '65세 이상'])
pop
```

```python
pop['소멸 비율'] = pop['20 - 39세', '여자'] / pop['65세 이상', '합계'] 
```

```python
pop['소멸 위기 지역'] = pop['소멸 비율'] < 0.5
pop
```

```python
pop[pop['소멸 위기 지역'] == True].index.get_level_values(1)
```
