```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0,10,0.01)
y = 3*t + 5

plt.figure( figsize = (12,8))
plt.plot(t,y)
plt.xticks(color = 'w')
plt.yticks(color = 'w')

plt.show()
```

```python
''' 직선을 찾는 재미를 주기 위해 노이즈를 살짝 줍니다!'''
y_noise = y + np.random.rand(len(y))
# 왜 len이 붙어야 하는지는 나중에 생각해보자
plt.figure( figsize = (12,8))
plt.plot(t, y_noise)
plt.xticks(color = 'w')
plt.yticks(color = 'w')
plt.show()
```

```python
''' 이제 직선을 찾아봅시다 
polyfit 명령으로 쉽게 찾을 수 있습니다. '''
fp1 = np.polyfit(t, y_noise, 1)
print(fp1)
''' 그런데 여기서 끝나는 게 아니라 (여기서는 계수밖에 얻은 게 없으니까)
함수(function)으로 만들어 줘야 함
poly1d : A one-dimensional polynomial class 
를 사용해서 polynomial class를 만들어 주면 됨. 폴리노미얼 클래스는 다항식으로 표현되면서 또 코드에서 
함수로 사용될 수도 있다.
'''
f1 = np.poly1d(fp1)
f1
```

```python
''' polynomial class : 다항식 연산이 가능한 numpy의 클래스 '''
print(np.poly1d([1,1])  + np.poly1d([1,-1]))
np.poly1d([1,1])  * np.poly1d([1,-1])
```

```python
plt.figure( figsize = (15,12) )
plt.plot(t, y_noise , label = 'noise', color = 'y')
plt.plot(t, y , ls = 'dashed', lw=3 , color = 'b', label ='original')
plt.plot(t, f1(t),  lw = 2 , color = 'r', label = 'polyfit')
plt.xticks(color = 'w')
plt.yticks(color = 'w')
# f1는 계수로 이뤄진 다항식의 클래스니까 실제 함수로 만들어주려면 f1(t)
plt.grid()
plt.legend()
plt.show()
```
