# 날짜별로 돌아가면서 크롤링

```python
from tqdm import tqdm

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date='

movie_date = []
movie_title = []
movie_point = []

for today in tqdm(date):
    res = requests.get(url + str(today)[:10])
    # pandas의 timestamps는 str을 붙여줘야 문자열이 돼요
    html = BeautifulSoup(raw, 'html.parser')
    
    end = html.select('td.title div.tit5 a')
    num = html.select('td.point')
                          
    movie_date.extend( [str(today)[:10] for n in range(0,len(end))])
    movie_title.extend( [name['title'] for name in end] )                          
    movie_point.extend( [point.text for point in num ] )                          
```

```python
movie_date
```

```python
len(movie_title)
```

```python
movie_point = list(map(float,movie_point))
movie_point
```

```python
movie = pd.DataFrame({
    'date':movie_date,
    'title':movie_title,
    'point':movie_point
})
movie.head(15)
```

```python
# 날짜는 너무 많으니까 영화별로 점수의 합산으로 보고 싶어! 그러면 pivot_table을 쓰겠지?
import numpy as np

movie_unique = pd.pivot_table(movie, index=['title'])
movie_unique
```

```python
movie_best = movie_unique.sort_values(by='point', ascending = False)
movie_best
```
