# 서울시 인구수와 구별 CCTV
 PANDAS로 텍스트 파일과 엑셀 읽기

```python
''' 엑셀 파일 읽기 '''
import pandas as pd # 모든 라이브러리는 클래스다!
                    # 그러므로 모든 함수는 타입함수(메소드) .head('인수 있으면 여기') 형태

CCTV_18 = pd.read_excel('CCTV_11~18.xlsx') # pd클래스의 메소드
CCTV_18.head() # 처음 5행까지만 보여준다.
```

```python
''' 열 인덱스'''
CCTV_18.columns 
# 메소드가 아니고 속성
# CCTV_Seoul.columns[0]
```

```python
''' "기관명" 이름을 "구별" 이름으로'''
CCTV_18.rename(columns=
    {CCTV_18.columns[0]: '구별'
}, inplace = True)
CCTV_18.head()
```

```python
''' 2번째 행까지 헤더가 있고, usecols를 써서 읽고싶은 열만 읽기 '''
pop_18 = pd.read_csv('pop18.txt',delimiter='\t', header = 2, usecols = [1,3,6,9,13], thousands = ',')
pop_18.head()
```

```python
''' 이름이 계, 계.1... 이렇게 나와있어서 알아보기 힘들다.'''
pop_18.rename(columns ={
    pop_18.columns[0] : '구별',
    pop_18.columns[1] : '인구수',
    pop_18.columns[2] : '한국인',
    pop_18.columns[3] : '외국인',
    pop_18.columns[4] : '고령자'
}, inplace = True)
pop_18.head()
```

최근 3년간의 CCTV 증가율

```python
CCTV_18.sort_values(by='소계',inplace=True, ascending=False)
CCTV_18.head()
```

```python
CCTV_18.iloc[:,2:-3].head()
```

```python
# 2015년 이전까지 총합
pop_15 = CCTV_18.iloc[:,2:-3].apply(sum, axis = 1)
pop_15.head()
```

```python
# 2016년 이후 총합
pop16_18 = CCTV_18.iloc[:,-3:].apply(sum, axis = 1)
pop16_18.head()
```

```python
CCTV_18['최근 3년 CCTV 증가율'] = pop_15 / pop16_18 * 100
```

```python
CCTV_18.head()
```

```python
pop_18.head()
```

```python
pop_18['외국인비율'] = pop_18['외국인'] / pop_18['인구수'] * 100 
pop_18['고령자비율'] = pop_18['고령자'] / pop_18['인구수'] * 100 
pop_18.head()
```

```python
result = pd.merge(CCTV_18, pop_18, on = '구별')
result
```

```python
result = pd.merge(CCTV_18[['구별','소계','최근 3년 CCTV 증가율']], pop_18, on = '구별')
result.head()
```

```python
import numpy as np
np.corrcoef(result['인구수'],result['CCTV 개수'])
```

## Regression
fp1 = np.polyfit( x축, y축, 차수): 데이터의 다항식 근사, 결과는 계수 튜플 <br>
f1 = np.poly1d( fp1(np.polyfit의 결과) ): 계수 튜플을 실제 함수로. y = f(x) 에서 f만 구현 <br>
plt.plot()에서 plot(x축,f1(x축)) 으로 실제 f 에 x를 넣어서 y를 얻어야 함


```python
# 양의 상관계수가 나온 인구수 - CCTV 개수 부터
fp1 = np.polyfit(result['인구수'] , result['CCTV 개수'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(1e5,7e5,100)
```

```python
plt.scatter(result['인구수'], result['CCTV 개수'], s = 50)
plt.plot(fx, f1(fx), color = 'r', ls = 'dashed', label = 'polyfit' )

plt.xlabel('인구수')
plt.ylabel('CCTV 개수')

plt.grid()
plt.show()
```

## Regression에서 현저히 먼 이상치들은 이름이 적혀지고, 멀어질 수록 색깔에 변화를 주기
오차 : 절대값(CCTV 개수 - f1(인구수)), 직선(f1)으로부터의 오차이기 떄문 <br>
colormap 파라미터는 c = 숫자 리스트

```python
result['오차'] = np.abs(result['CCTV 개수'] - f1(result['인구수']) )
```

```python
er_sort = result.sort_values(by = '오차', ascending = False )
er_sort.head()
```

그래프 위에 이름 붙이기, 오차가 심한 10개 구 선택

```python
plt.scatter( er_sort['인구수'] , er_sort['CCTV 개수'],
           c = er_sort['오차'] , s = 50)
plt.plot(fx,f1(fx), ls='dashed',lw=3 , color = 'g')


for i in range(10):
    plt.text(er_sort['인구수'][i]*1.02 , er_sort['CCTV 개수'][i]*0.98,
             # x 값으로부터 1.02, y값으로부터 0.98만큼 떨어진 곳에 text를 표시
             er_sort.index[i] , fontsize = 15)
             # 뽑아낼 문자열

plt.show()
```
