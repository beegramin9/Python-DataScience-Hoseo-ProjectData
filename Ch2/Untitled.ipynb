```python
import json
geo_path = '02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open (geo_path, encoding= 'utf-8'))
```

```python
import pandas as pd
crime_norm = pd.read_csv('02 crime_in_Seoul_final.csv', encoding = 'utf-8')
crime_norm.set_index('구별',inplace = True)
crime_norm.head(15)
```

```python
crime_norm.index
```

```python
import folium
map_murder = folium.Map( location = [37.5502, 126.982] , zoom_start = 11,
                tiles = 'Stamen Toner')
map_murder.choropleth(geo_data = geo_str,
                data = crime_norm['살인'],
                columns = [crime_norm.index, crime_norm['살인'] ],
                fill_color = 'PuRd',
                key_on = 'feature.id')
map_murder
```

```python
map_rape = folium.Map( location = [37.5502, 126.982] , zoom_start = 11,
                tiles = 'Stamen Toner')
map_rape.choropleth(geo_data = geo_str,
                data = crime_norm['강간'],
                columns = [crime_norm.index, crime_norm['강간'] ],
                fill_color = 'PuRd',
                key_on = 'feature.id')
map_rape
```

```python
map_overall = folium.Map( location = [37.5502, 126.982] , zoom_start = 11,
                tiles = 'Stamen Toner')
map_overall.choropleth(geo_data = geo_str,
                data = crime_norm['범죄'],
                columns = [crime_norm.index, crime_norm['범죄'] ],
                fill_color = 'PuRd',
                key_on = 'feature.id')
map_overall
```

- 근데 여기서 우리가 하나 간과한 게 있죠! 범죄 발생 수만 보는게 아니라 인구수에 비례해서 보아야지
맞다고 할 수 있겠지요

```python
rate_crime = crime_norm['살인'] / crime_norm['인구수'] * 1000000 # 색깔의 차이를 크게 보여주겠지

map_rate = folium.Map( location = [37.5502, 126.982] , zoom_start = 11,
                tiles = 'Stamen Toner')
map_rate.choropleth(geo_data = geo_str,
                data = crime_norm['범죄'],
                columns = [crime_norm.index, rate_crime ],
                fill_color = 'PuRd',
                key_on = 'feature.id')
map_rate

''' 그런데 중구와 종로구의 범죄율이 엄청 높아졌습니다! 아무래도 거주 인구가 적고, 관광지여서 그런 게 아닐지... '''
```

```python
station_name = []

crime_police = pd.read_csv('02. crime_in_Seoul.csv', thousands = ',', encoding = 'euc-kr')

for name in crime_police['관서명']:
    station_name.append('서울'+str(name[:-1] + '경찰서'))
    
station_name
```

```python
station_address = []
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name, language = 'ko')
    station_address.append(tmp[0].get('formatted_address'))
    
    tmp_loc = tmp[0].get('geometry')
    
    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])    
    
    print( name + '-->' + tmp[0].get('formatted_address'))
```
