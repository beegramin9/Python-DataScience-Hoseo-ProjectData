```python
from sklearn.impute import SimpleImputer
''' 2018년에 개정되었음 
더 이상 from sklearn.preprocessing import imputer 라고 안 씁니다. '''

df = data.copy()
# 이래야 복사된다고 했죠? copy()를 안 쓰면 data까지 바뀐다고 했어요
df.drop([
    'Name',
    'Sex',
    'Ticket',
    'Cabin',
    'Embarked'
], axis = 1, inplace = True)
# 숫자 데이터만 보기 위해서 텍스트로 되어있는 데를 없앤 거에요
# axis = 1 이니까 x축을 전부 없애는 거죠!
df.head(10)
```

```python
imputer = SimpleImputer(strategy='median')
''' strategy의 옵션 값으로는 'mean', 'medain', 'most_frequent' 가 있습니다.
imput는 attribute, ascribe 와 같은 뜻으로 ~의 탓으로 돌리다 인데요... 궁금하네용
빈 공간을 위 3가지 옵션 중 하나로 대체하겠다 라는 뜻 같네'''

''' 빈 공간 Nan 들을 채우기 위해 fit을 사용하면 된다. 
그러면 채워집니다!'''
imputer.fit(df)
print(imputer.statistics_)
'''
SimpleImputer 클래스에서 Pandas의 describe() 와 같은 역할을 합니다.


'''
```

```python
output = imputer.transform(df)
''' array 형태로 만들어 줍니다. ''' 
```

```python
output = pd.DataFrame(output, columns= df.columns , index = list(df.index.values) )
output.describe()
''' 이러면 새로운 DataFrame이 만들어지면서 빈 넘버가 다 채워져있는 걸 알 수 있죠? 
그리고 Nan들은 전부 중앙값으로 바뀌어져 있겠죠 '''
```

```python
''' sklearn.preprocessing 패키지가 제공하는 MinMaxScaler/ StandardScaler를 활용해서 정규화 해보기 '''
# 우선 함수로 min-max scaler 가 뭐하는 놈인지부터 봅시다.
def norm(x):
    _max = x.max()
    _min = x.min()
    _denom = _max - _min
    return (x - _min) / _denom
# 네, 이런 일을 한다고 합니다.
```
