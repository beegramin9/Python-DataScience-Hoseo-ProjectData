```python
''' 구 이름이 겹치는데 어떻게 정리하지? 
Pandas의 pivot_table'''

import pandas as pd
import numpy as np

df = pd.read_excel('sales-funnel.xlsx')
df
```

```python
''' 위 데이터에서 Name 항목으로만 정렬할 때 pivot_table을 사용합니다.
그러면 Name 컬럼이 index가 되고 특별히 지정하지 않는다면 숫자형 데이터 컬럼만 남습니다.
중복된 Name들의 항목은 하나로 합쳐주고 value는 평균값이 된다.
'''
name_pivot = pd.pivot_table(df, index = ['Name'])
''' 인덱스를 여러 개 지정할 수 있기 때문에 리스트로 되어있습니다. '''
print(name_pivot)
multi_pivot = pd.pivot_table(df, index = ['Name','Rep','Manager'])
print(multi_pivot)
''' 특정 value만 지정해서 나타나게 할 수 있습니다. '''
certain_value = pd.pivot_table(df, index = ['Manager','Rep'], values = 'Price')
certain_value # Values의 값은 평균치로 나옵니다!
```

```python
''' 근데 나는 평균치 내는 거 싫고 다 더한 값이 좋다! 라고 하면 
, aggfun = np.sum '''
agg_func = pd.pivot_table(df, index = ['Manager','Rep'], values = 'Price' , aggfunc = np.sum)
agg_func
```

```python
pd.pivot_table(df, index = ['Manager','Rep','Product'] , values = ['Price','Quantity'],
              aggfunc=[np.sum, np.mean])
```

```python
crime_anal_raw = pd.read_csv('02. crime_in_Seoul_include_gu_name.csv', encoding = 'utf-8')
crime_anal_raw
# unnamed는 그냥 인덱스네
del crime_anal_raw['Unnamed: 0']
```

```python
''' pivot_table을 이용해서 원 데이터를 관서별에서 구별로 바꾸면? '''
crime = pd.pivot_table(crime_anal_raw, index = ['구별'] , aggfunc = np.sum)
crime
# 평균을 내 버리면 안되고 다 합쳐야겠지?
```

```python
crime['강간 검거율'] = crime['강간 검거'] / crime['강간 발생'] *100
crime['강도 검거율'] = crime['강도 검거'] / crime['강도 발생'] *100
crime['살인 검거율'] = crime['살인 검거'] / crime['살인 발생'] *100
crime['절도 검거율'] = crime['절도 검거'] / crime['절도 발생'] *100
crime['폭력 검거율'] = crime['폭력 검거'] / crime['폭력 발생'] *100

del crime['강간 검거']
del crime['강도 검거']
del crime['살인 검거']
del crime['절도 검거']
del crime['폭력 검거']

crime
```

```python
crime['강간 검거율'] = round( crime['강간 검거율'], 2) 
crime['강도 검거율'] = round( crime['강도 검거율'], 2) 
crime['살인 검거율'] = round( crime['살인 검거율'], 2)
crime['절도 검거율'] = round( crime['절도 검거율'], 2)
crime['폭력 검거율'] = round( crime['폭력 검거율'], 2)

crime.rename( columns = {
    '강간 발생' : '강간',
    '강도 발생' : '강도',
    '살인 발생' : '살인',
    '절도 발생' : '절도',
    '폭력 발생' : '폭력'
} , inplace = True)





crime
```

```python
''' 강간은 100단위, 강도와 살인은 10단위, 절도와 폭력은 1000단위라서 비교하기가 힘들다! 
이를 변수들의 척도(Scale) 이 다르다 고 한다.
그래서 정규화 (내가 아는 그 Z가 맞다. Normalization 을 하자 )
'''

from sklearn import preprocessing

col = [
    '강간',
    '강도',
    '살인',
    '절도',
    '폭력'
]

x = crime[col].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform( x.astype(float) )
''' fit_ransform할 대상을 넣는데, 숫자 타입을 float로 만든다는 거야.
요 fit_transform을 하게 되면 대상은 array로 나오는데, 그래서 밑에서
columns와 index를 crime의 걸 빌려서 다시 해주는거야.
'''
crime_norm = pd.DataFrame( x_scaled,
                         columns = col,
                         index = crime.index )

''' 여기서는 정규화 할 부분만 바꿔줬으니까, col만 해준거지 
'''

col2 = [
    '강간 검거율',
    '강도 검거율',
    '살인 검거율',
    '절도 검거율',
    '폭력 검거율'
]

crime_norm[col2] = crime[col2]
''' 그리고 여기서 정규화 안 시킨애들을 새로운 crime_norm 에다가 박아넣어준 거야 '''


crime_norm
```

```python
''' 위의 코드가 대체 무슨 소리지?
https://rfriend.tistory.com/270
"최소값(min)과 최대값(max)를 사용해서 0~1 사이의 범위(range)로 데이터를 표준화 해준다. "
'''
x = np.array([
[ 10, -10,1],
[ 5, 0,2],        
[ 0, 10,3]
])
x
```

```python
''' 1. 최소/최대값을 구해서 수동으로 '0~1' 범위로 변환 
'''
x_MinMax = (x -x.min(axis = 0)  /  x.max(axis =0) - x.min(axis = 0))
x_MinMax
```

```python
''' 2. sklearn.preprocessing.MinMaxScaler() method를 이용해서 0~1 변환 
'''
min_max_scaler = preprocessing.MinMaxScaler()
x_MinMax_train = min_max_scaler.fit_transform(x)
x_MinMax_train
```

```python
''' 새로운 테스트 셋 '''
y = np.array([
    [9, -10 ,1],
    [5, -5  ,3],
    [1, 0, 5]
])
y_MinMax = min_max_scaler.transform(y)
y_MinMax
```

```python
''' 그리고 CCTV 결과 csv를 붙여보겠습니다. '''
result_CCTV = pd.read_csv('../Ch1_cctv_district/01. CCTV_result.csv', index_col = '구별')
''' 요기있는 index_col = '구별' 은 result_CCTV.set_index('구별') 이랑 똑같아요!
'''
result_CCTV.head(10)

crime_norm[['인구수', 'CCTV']] = result_CCTV[['인구수', 'CCTV 개수']]
crime_norm.head(10)
```

```python
''' 이제 발생 건수의 합을 "범죄" 항목으로 두고 합하겠습니다. '''
col = ['강간',
   '강도',
   '살인',
   '절도',
   '폭력'
]
crime_norm['범죄'] = np.sum(crime_norm[col] , axis = 1)
''' 행렬이 있는데 x축(행방향) 으로 합쳐주는 것입니다.
정규화를 하지 않았다면 몇 천 건의 절동 수십 건의 살인 비중이 애매했겠지만 정규화를 통해 그 건 알아봅기 쉽게 되었습니다.'''
crime_norm.head(10)

```
